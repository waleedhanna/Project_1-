```python
# Read CSV
df_raw = pd.read_csv('gapminder.tsv', sep='\t')
print(df_raw.shape)
df_raw.head()
df_raw.info()
```

```python
#df = df_raw.set_index(['year'])
```

```python
df= df_raw.groupby(['country']).mean()
```

```python
df.head(10)
```

```python
df['pop'] = df['pop'] / 1000000
df['year'] = df['year']
df.head()
```

```python
df.columns
```

```python
pop = df['pop'].tolist()
pop[:10]
```

```python
gdp_cap = df['gdpPercap'].tolist()
gdp_cap[:10]
```

```python
life_exp = df['lifeExp'].tolist()
life_exp[:10]
```

```python
year = df['year'].tolist()
year [:10]
```

```python
#   Visualization
```

```python
# Create scatter plot
plt.scatter(gdp_cap, life_exp)



# Put the x-axis on a logarithmic scale
plt.xscale('log')

# Show plot
plt.show()

```

```python
# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log') 
# Store pop as a numpy array: np_pop
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop * 2
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Strings set to x, y and title variables 
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'POP GDP VS life Expectancy'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)
# Definition of tick_val and tick_lab
tick_val = [1000,10000,100000]
tick_lab = ['1k','10k','100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)
# After customizing, display the plot
# Add grid() call
plt.grid(True)

fig = plt.gcf()
fig.set_size_inches(10, 8)



plt.show()


# Show the plot
plt.show()
```

```python
df_1 = df_raw[['continent',  'lifeExp',  'gdpPercap', 'pop']]
```

```python
df_cont = df_1.groupby(['continent'])
df_cont_group=df_cont.mean()
df_cont_group
```

```python
color = ['red', 'green', 'blue', 'yellow', 'black']
```

```python
#df_raw['color'] = df_raw.apply(color_map, axis=1)
df_raw.head()
 
```

```python

plt.scatter(gdp_cap, life_exp)
plt.xscale('log') 
# Store pop as a numpy array: np_pop
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop * 3
plt.scatter(gdp_cap, life_exp, s = np_pop)



plt.scatter(x = df_cont_group["gdpPercap"], y = df_cont_group["lifeExp"], c = color, alpha = 0.8)

# Labels
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('Continent GDP VS life Expectancy')
plt.legend=df_cont_group.index.values
# Definition of tick_val and tick_lab
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Add grid() call
plt.grid(True)

fig = plt.gcf()
fig.set_size_inches(10, 8)

# Show the plot
plt.show()
```
